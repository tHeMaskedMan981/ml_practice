```python
#import the necessary module
from sklearn import preprocessing
# create the Labelencoder object
le = preprocessing.LabelEncoder()
#convert the categorical columns into numeric
sales_data['BusinessTravel'] = le.fit_transform(sales_data['BusinessTravel'])
sales_data['Department'] = le.fit_transform(sales_data['Department'])
sales_data['EducationField'] = le.fit_transform(sales_data['EducationField'])
sales_data['Gender'] = le.fit_transform(sales_data['Gender'])
sales_data['JobRole'] = le.fit_transform(sales_data['JobRole'])
sales_data['OverTime'] = le.fit_transform(sales_data['OverTime'])
sales_data['MaritalStatus'] = le.fit_transform(sales_data['MaritalStatus'])

#convert the categorical columns into numeric
data_test['BusinessTravel'] = le.fit_transform(data_test['BusinessTravel'])
data_test['Department'] = le.fit_transform(data_test['Department'])
data_test['EducationField'] = le.fit_transform(data_test['EducationField'])
data_test['Gender'] = le.fit_transform(data_test['Gender'])
data_test['JobRole'] = le.fit_transform(data_test['JobRole'])
data_test['OverTime'] = le.fit_transform(data_test['OverTime'])
data_test['MaritalStatus'] = le.fit_transform(data_test['MaritalStatus'])

#display the initial records
sales_data.head()
data_test.head()
```

```python
# select columns other than 'Attrition'
cols = [col for col in sales_data.columns if col not in ['Attrition']]
# dropping the 'Attrition' columns
data_train = sales_data[cols]
#assigning the Oppurtunity Result column as target
target_train = sales_data['Attrition']
data_train.head()
```

```python
# import the necessary module
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
#create an object of the type GaussianNB
gnb = GaussianNB()
data_train.dtypes
#train the algorithm on training data and predict using the testing data
pred1 = gnb.fit(data_train, target_train).predict(data_test)
print(pred1)
#print the accuracy score of the model
# print("Naive-Bayes accuracy : ",accuracy_score(target_test, pred, normalize = True))
```

```python
#import the necessary modules
from sklearn.svm import LinearSVC
#create an object of type LinearSVC
svc_model = LinearSVC(random_state=0)
#train the algorithm on training data and predict using the testing data
pred2 = svc_model.fit(data_train, target_train).predict(data_test)
#print the accuracy score of the model
# print("LinearSVC accuracy : ",accuracy_score(target_test, pred, normalize = True))
```

```python
#import necessary modules
from sklearn.neighbors import KNeighborsClassifier
#create object of the lassifier
neigh = KNeighborsClassifier(n_neighbors=3)
#Train the algorithm
neigh.fit(data_train, target_train)
# predict the response
pred3 = neigh.predict(data_test)
# evaluate accuracy
# print ("KNeighbors accuracy score : ",accuracy_score(target_test, pred))
```
